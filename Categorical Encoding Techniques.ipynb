### Replace method

Could be used for Ordinal data. Not a good idea for Nominal feature especially if it has lot of categories

```python
df1 = data.select_dtypes('object').copy()
```

```python
df1
```

```python
df1['salary'].value_counts()
```

```python
replacesal = {'salary':{'<=50K': 0,
                        '>50K': 1}}

df1.replace(replacesal)
```

### OrdinalEncoder sklearn

```python
df3 = data.select_dtypes('object').copy()
```

```python
df3.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordencoder = OrdinalEncoder()
df3['Salary_encode'] = ordencoder.fit_transform(df3[['salary']])
df3[['salary','Salary_encode']].head()
```

### One Hot Encoder sklearn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df3.head()
```

```python
onehot = OneHotEncoder()
oe_results = onehot.fit_transform(df3[['race']])
mydf = pd.DataFrame(oe_results.toarray(),columns = onehot.categories_)
```

```python
mydf.head()
```

```python
finaldf = df3.join(mydf)
```

```python
finaldf
```

```python
onehot = OneHotEncoder()
oe_results = onehot.fit_transform(df3[['occupation','education']])
mydf = pd.DataFrame(oe_results.toarray(),index = df3.index)
```

### Assumption 2:independent, identically distributed residual errors

### Test for independence of residual errors

```python
model_expr = 'PE ~ AT + V + AP + RH'
```

```python
y, X = dmatrices(model_expr, df, return_type='dataframe')
```

```python
y
```

```python
X
```

```python
mask = np.random.rand(len(X)) < 0.8
X_train = X[mask]
y_train = y[mask]
X_test = X[~mask]
y_test = y[~mask]
```

```python
import statsmodels.api as sm
olsr_results = sm.OLS(y_train, X_train).fit()
print('Training completed')

print(olsr_results.summary())
```

```python
olsr_predictions = olsr_results.get_prediction(X_test)
```

```python
prediction_summary_frame = olsr_predictions.summary_frame()
print(prediction_summary_frame)
```

```python
resid = y_test['PE'] - prediction_summary_frame['mean']
```

```python
plt.xlabel('Predicted Power Output', fontsize=18)
plt.ylabel('Residual Error of Regression', fontsize=18)
plt.scatter(prediction_summary_frame['mean'], resid)
plt.show()
```

What identically distributed means is that residual error ε_i corresponding to the prediction for each data row, has the same probability distribution. If the distribution of errors is not identical, one cannot reliably use tests of significance such as the F-test for regression analysis or perform confidence interval testing on the predictions. Many of these tests depend on the residual errors being identically, and normally distributed.

### F-test for regression analysis

The F-test, when used for regression analysis, lets you compare two competing regression models in their ability to “explain” the variance in the dependent variable.

The F-test can be used to answer the following questions:
- Will you be able to improve your linear regression model by making it more complex i.e. by adding more linear regression variables to it?
- Or making it more simple (restricted)
